# Replace non-numerical values with the right code

```python
def replace_str(this_val):
    """ Replace non-numerical values of this_val by replacement code."""
    
    replacement_code = this_val
    
    if isinstance(this_val, basestring):
        if this_val == '..':
            print this_val + ' replaced by na_current_period'
            replacement_code = 'na_current_period'
        elif this_val == '..t':
            print this_val + ' replaced by terminated'
            replacement_code = 'terminated'
        elif [x for x in this_val if 't' in x]: #remove 't' from actual number value
            print this_val + ' num value extracted'
            num_only = this_val.split('t')[0].replace(',', '') #remove decimal placeholder comma!
            replacement_code = int( num_only )
            print num_only
        else:
            replacement_code = int(this_val)
    
    return replacement_code
```

```python
dir = '/home/nangini/PROJECTS/airports/data/'

#loop through regions via column numbers
for idx in range(0, len(colnums)-1):
    
    print regions_dict[regions[idx]]    #regions[idx].replace(" ", "")
    #Define file name based on region name
    file = open(dir + regions_dict[regions[idx]] + '_FuelSales.json', 'w')
        
    #Extract subset from df corresponding to regions (use column numbers)
    s1 = colnums[idx]
    s2 = colnums[idx+1]
    print s1, s2
    
    #Define empty df for this subset
    this_df = pd.DataFrame()
    this_df = df.iloc[:,s1:s2]
    
    #FORMAT
    #drop Gross sales
    this_df.drop(['Gross sales of gasoline'], inplace=True)
    #replace column names with first row names
    this_df = this_df.rename(columns=this_df.iloc[0])
    #make sure the column names are integers
    this_df.columns = this_df.columns.map(int)
    #remove first row
    this_df.drop(['year'], inplace=True)
    
    #Define array of years specific to this_df
    years = list(this_df)
    print years
    print ""
    
    file.write('{\n')
    file.write('  "fuelsales": [\n')

    #Loop over each year and extract gas, diesel, LPG
    for year in years:
        file.write('    { \n')       
        
        year = int(year)
        val_gas = this_df[year]['Net sales of gasoline']
        val_diesel = this_df[year]['Net sales of diesel oil']
        val_lpg = this_df[year]['Net sales of liquefied petroleum gas']
        
        #check for non-numerical str and replace with replacement code
        val_gas = replace_str(val_gas)
        val_diesel = replace_str(val_diesel)
        val_lpg = replace_str(val_lpg)
        
        file.write( '      "year": %d,\n' %( year ) )
        
        if isinstance(val_gas, basestring): #if str, write out a str
            file.write( '      "gas": "%s",\n' %( val_gas ) )
        else: #else write int
            file.write( '      "gas": %d,\n' %( val_gas ) )
            
        if isinstance(val_diesel, basestring):
            file.write( '      "diesel": "%s",\n' %( val_diesel ) )
        else:
            file.write( '      "diesel": %d,\n' %(val_diesel ) )
            
        if isinstance(val_lpg, basestring):
            file.write( '      "LPG": "%s"\n' %( val_lpg ) )
        else:
            file.write( '      "LPG": %d\n' %( val_lpg ) )
                
        file.write('    },\n')

    #remove last comma
    file.seek(-2, os.SEEK_END)
    file.truncate()

    file.write('\n  ]\n')
    file.write('}')

    file.close()
```
